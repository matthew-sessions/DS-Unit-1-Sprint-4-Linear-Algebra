## 1.1 Create a two-dimensional vector and plot it on a graph

```
import numpy as np
import matplotlib.pyplot as plt
two_dem = np.array([5,2])

plt.arrow(0,0,two_dem[0], two_dem[1],head_width=.3, head_length=.3)
plt.xlim(0,6)
plt.ylim(0,6)
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D

three_dem = np.array([5,2,3])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0,0,0,three_dem[0], three_dem[1], three_dem[2], length=1)
ax.set_xlim(0,6)
ax.set_ylim(0,6)
ax.set_zlim(0,6)
```
